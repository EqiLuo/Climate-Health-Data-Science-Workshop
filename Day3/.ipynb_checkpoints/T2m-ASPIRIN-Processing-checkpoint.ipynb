```python
# load packages
import xarray as xr
import numpy as np
import pandas as pd
from datetime import datetime as dt
```

```python
# load NetCDF file and print details
TempFilePath = '/home/p18q126/grp-ct/data/raw/climate/era5/WetBulbTempInputData/WetBulbTempInput-201[678]*.nc'
TempData = xr.open_mfdataset(TempFilePath)

# Define start and end dates
start_date = '2016-03-01'
end_date = '2018-07-01'

# Convert start and end dates to datetime objects
start_date_dt = pd.to_datetime(start_date)
end_date_dt = pd.to_datetime(end_date)

# filter dataset based on dates
TempDataFiltered = TempData.sel(time=slice(start_date_dt, end_date_dt))
TempDataFiltered
```

```python
# narrows the dataset using max extent of the points we want to intersect
# at the moment this is manual but can easily change to use min/max lat/lon automatically
TempDataFilteredNarrow = TempDataFiltered.sel(longitude=slice(-91.05234385534075, 80.02054906545357), latitude=slice(32.73045974192184, -15.780012098523654, ))
TempDataFilteredNarrow
```

```python
# load Tabular data to extract heat values
PointFilePath = '/home/p18q126/grp-ct/data/raw/surveys/ASPIRIN-Heat-Stress/ASPIRIN Heat Stress Clusters Listing (2023.11.29).csv'
PointData = pd.read_csv(PointFilePath)

# Split the 'Location' column into 'Latitude' and 'Longitude' columns
PointData[['Latitude', 'Longitude']] = PointData['Latitude, Longitude Decimal Degrees'].str.split(',', expand=True)

# Drop the original 'Location' column
PointData.drop(columns=['Latitude, Longitude Decimal Degrees'], inplace=True)

# Rename columns
new_column_names = {'Number Randomized': 'NumRandom', 'Date First Participant Randomized': 'DateRandom', 'Date Last Participant Delivered': 'DateDelivered'}
PointData.rename(columns=new_column_names, inplace=True)

# make lat and lon a float
PointData['Latitude'] = PointData['Latitude'].astype(float)
PointData['Longitude'] = PointData['Longitude'].astype(float)
PointData
```

```python
# set latitude and longitude variables, as well as name for cluster ID
lat = PointData['Latitude']
lon = PointData['Longitude']
name = PointData['Cluster']
```

```python
# for loop, loop through each lat/lon.cluster and extract the value from the dataset using the nearest approach
# concat these in to a new dataframe
Newdf = pd.DataFrame([])
dfs_to_concat = []

for i, j, k in zip(lat, lon, name):
    dsloc = TempDataFilteredNarrow.sel(latitude=i, longitude=j, method='nearest')
    DT = dsloc.to_dataframe()

    # Insert the name of the station with preferred column title:
    DT.insert(loc=0, column="Cluster", value=k)
    dfs_to_concat.append(DT)

Newdf = pd.concat(dfs_to_concat, sort=True)
```

```python
# make the dataframe time column and format correctly
Newdf.reset_index(inplace=True)

Newdf['time'] = pd.to_datetime(Newdf['time'])

# Format the time column to your desired format
Newdf['time'] = Newdf['time'].dt.strftime('%Y-%m-%d-%H')
Newdf['t2m_C'] = Newdf['t2m'] - 273.15
Newdf.drop(['d2m', 'sp', 't2m'], axis=1, inplace=True)
Newdf
```

```python
# create a pivot table so we have cluster id and then columns for each time point
pivot_df = Newdf.pivot_table(index=['latitude', 'longitude', 'Cluster'], columns='time', values='t2m_C', aggfunc='first')

# Reset the index to make 'latitude', 'longitude', and 'cluster' as regular columns
pivot_df.reset_index(inplace=True)
pivot_df
```

```python
# Merge the original DataFrame information into pivot_df
merged_df = pd.merge(pivot_df, PointData[['Site', 'Cluster', 'NumRandom', 'DateRandom', 'DateDelivered', 'Latitude', 'Longitude']], on='Cluster', how='left')
# Drop duplicate latitude and longitude columns
merged_df.drop(columns=['latitude', 'longitude'], inplace=True)

column_order = ['Site', 'Cluster', 'NumRandom', 'DateRandom', 'DateDelivered', 'Latitude', 'Longitude'] + list(pivot_df.columns[3:])  # Assuming the time columns start from the 4th column
merged_df = merged_df[column_order]

# Display the merged DataFrame
merged_df
```

```python
#save as csv
merged_df.to_csv('/home/p18q126/grp-ct/data/raw/surveys/ASPIRIN-Heat-Stress/ASPIRIN-T2mC-Output.csv', index=False)
```

```python
# calculate daily max, min and range
Newdf['time'] = pd.to_datetime(Newdf['time'])
Newdf['date'] = Newdf['time'].dt.date

daily_stats = Newdf.groupby(['Cluster', 'latitude', 'longitude', 'date']).agg(
    max_t2mC=('t2m_C', 'max'),
    min_t2mC=('t2m_C', 'min'),
    range_t2mC=('t2m_C', lambda x: x.max() - x.min())
).reset_index()
daily_stats
```

```python
daily_stats = daily_stats.drop(['latitude', 'longitude'], axis=1) # we are removing this because this is the lat/lon for the pixel, rather than the camp. We will add back in the camp lat/lon
daily_stats = pd.merge(daily_stats, PointData, on='Cluster', how='inner')
daily_stats
```

```python
daily_stats.to_csv('/home/p18q126/grp-ct/data/raw/surveys/ASPIRIN-Heat-Stress/ASPIRIN-T2m_C-Daily-Output.csv', index=False)
```
